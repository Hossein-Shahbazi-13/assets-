```python
# Ù†ØµØ¨ Ú©ØªØ§Ø¨Ø®Ø§Ù†Ù‡â€ŒÙ‡Ø§ÛŒ Ù…ÙˆØ±Ø¯ Ù†ÛŒØ§Ø² (Ø§Ú¯Ø± Ù„Ø§Ø²Ù… Ø¨ÙˆØ¯)
!pip install scikit-learn matplotlib

# Ú©ØªØ§Ø¨Ø®Ø§Ù†Ù‡â€ŒÙ‡Ø§
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, LSTM, Dense, Dropout, BatchNormalization, Reshape
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
from google.colab import drive

# Ø§ØªØµØ§Ù„ Ø¨Ù‡ Ú¯ÙˆÚ¯Ù„ Ø¯Ø±Ø§ÛŒÙˆ
drive.mount('/content/drive')

# Ù…Ø³ÛŒØ± ÙØ§ÛŒÙ„â€ŒÙ‡Ø§ÛŒ CSV
file_paths = [
    '/content/drive/MyDrive/NUSW-NB15/UNSW-NB15_1.csv',
    '/content/drive/MyDrive/NUSW-NB15/UNSW-NB15_2.csv',
    '/content/drive/MyDrive/NUSW-NB15/UNSW-NB15_3.csv',
    '/content/drive/MyDrive/NUSW-NB15/UNSW-NB15_4.csv'
]



# Ø®ÙˆØ§Ù†Ø¯Ù† Ùˆ Ø§Ø¯ØºØ§Ù… Ø¯Ø§Ø¯Ù‡â€ŒÙ‡Ø§
dfs = [pd.read_csv(file, header=None) for file in file_paths]
data = pd.concat(dfs, axis=0)

# Ø­Ø°Ù Ù…Ù‚Ø§Ø¯ÛŒØ± Ø®Ø§Ù„ÛŒ
data.dropna(inplace=True)

# Ø§ØµÙ„Ø§Ø­ Ù†Ø§Ù… Ø³ØªÙˆÙ†â€ŒÙ‡Ø§
column_names = ['srcip', 'sport', 'dstip', 'dsport', 'proto', 'state', 'dur', 'sbytes', 'dbytes',
                'sttl', 'dttl', 'sloss', 'dloss', 'service', 'Sload', 'Dload', 'Spkts', 'Dpkts',
                'swin', 'dwin', 'stcpb', 'dtcpb', 'smeansz', 'dmeansz', 'trans_depth',
                'res_bdy_len', 'Sjit', 'Djit', 'Stime', 'Ltime', 'Sintpkt', 'Dintpkt',
                'tcprtt', 'synack', 'ackdat', 'is_sm_ips_ports', 'ct_state_ttl', 'ct_flw_http_mthd',
                'is_ftp_login', 'ct_ftp_cmd', 'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ltm',
                'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'attack_cat', 'label']
data.columns = column_names

# ÙÙ‚Ø· ØªØ¹Ø¯Ø§Ø¯ Ù…Ø´Ø®ØµÛŒ Ø§Ø² Ø¯Ø§Ø¯Ù‡â€ŒÙ‡Ø§ Ø¨Ø±Ø§ÛŒ Ø¢Ù…ÙˆØ²Ø´ Ø§ÙˆÙ„ÛŒÙ‡ (Ù…Ø«Ù„Ø§Ù‹ 300 Ù‡Ø²Ø§Ø± Ø±Ú©ÙˆØ±Ø¯)
data = data.sample(n=10000, random_state=42)

# Ø­Ø°Ù Ø³ØªÙˆÙ†â€ŒÙ‡Ø§ÛŒ ØºÛŒØ±Ù…ÙÛŒØ¯
data.drop(columns=['srcip', 'dstip'], inplace=True)

# ØªØ¨Ø¯ÛŒÙ„ ÙˆÛŒÚ˜Ú¯ÛŒâ€ŒÙ‡Ø§ÛŒ Ù…ØªÙ†ÛŒ Ø¨Ù‡ Ø¹Ø¯Ø¯ÛŒ
categorical_cols = ['proto', 'service', 'state']
data = pd.get_dummies(data, columns=categorical_cols)

# Ø¬Ø¯Ø§ Ú©Ø±Ø¯Ù† ÙˆÛŒÚ˜Ú¯ÛŒâ€ŒÙ‡Ø§ Ùˆ Ø¨Ø±Ú†Ø³Ø¨
X = data.drop(columns=['label', 'attack_cat'])
y = data['attack_cat']

# ØªØ¨Ø¯ÛŒÙ„ Ø¨Ø±Ú†Ø³Ø¨ Ø¨Ù‡ Ø¹Ø¯Ø¯ÛŒ
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Ù†Ø±Ù…Ø§Ù„â€ŒØ³Ø§Ø²ÛŒ
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ØªÙ‚Ø³ÛŒÙ… Ø¨Ù‡ Ø¯Ø§Ø¯Ù‡ Ø¢Ù…ÙˆØ²Ø´ Ùˆ Ø¢Ø²Ù…ÙˆÙ†
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_categorical, test_size=0.2, random_state=42)

# ØªØºÛŒÛŒØ± Ø´Ú©Ù„ Ø¨Ø±Ø§ÛŒ Conv1D
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

# Ø³Ø§Ø®Øª Ù…Ø¯Ù„ CNN + LSTM
input_layer = Input(shape=(X_train.shape[1], 1))

x = Conv1D(128, kernel_size=3, activation='relu')(input_layer)
x = BatchNormalization()(x)
x = MaxPooling1D(pool_size=2)(x)

x = Conv1D(256, kernel_size=3, activation='relu')(x)
x = BatchNormalization()(x)
x = MaxPooling1D(pool_size=2)(x)

x = Reshape((x.shape[1], 256))(x)

x = LSTM(128, return_sequences=True)(x)
x = LSTM(64, return_sequences=False)(x)

x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.5)(x)

output_layer = Dense(y_categorical.shape[1], activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output_layer)

# Ú©Ø§Ù…Ù¾Ø§ÛŒÙ„ Ù…Ø¯Ù„
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Ø¢Ù…ÙˆØ²Ø´ Ù…Ø¯Ù„
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=50,
                    batch_size=128,
                    callbacks=[early_stop])

# Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ Ù†Ù‡Ø§ÛŒÛŒ
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\nâœ… Test Accuracy: {accuracy*100:.2f}%")

# Ú¯Ø²Ø§Ø±Ø´ Ø·Ø¨Ù‚Ù‡â€ŒØ¨Ù†Ø¯ÛŒ
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

print("\nğŸ§¾ Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_))

# Confusion Matrix
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.heatmap(confusion_matrix(y_true, y_pred_classes), annot=True, fmt='d',
            xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Ù†Ù…ÙˆØ¯Ø§Ø± Ø¯Ù‚Øª Ùˆ Ø®Ø·Ø§
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss")
plt.legend()
plt.show()
```
